```python
import requests
import json

def getPoemUrls(link, poemUrls):
    url = "https://ganjgah.ir/api/ganjoor/cat?url=" + link + "&poems=true&mainSections=false"

    payload={}
    headers = {
      'Accept': 'application/json'
    }
    response = requests.request("GET", url, headers=headers, data=payload)
    data = json.loads(response.text)
    for poem in data["cat"]["poems"]:
        poemUrls.append(link + "/" + poem["urlSlug"])
    for book in data["cat"]["children"]:
        getPoemUrls(link + "/" + book["urlSlug"], poemUrls)
        
def getPoemByUrl(link):
    url = "https://ganjgah.ir/api/ganjoor/poem?url=" + link + "&catInfo=true&catPoems=false&rhymes=true&recitations=true&images=true&songs=true&comments=true&verseDetails=true&navigation=true"
    payload={}
    headers = {
      'Accept': 'application/json'
    }
    response = requests.request("GET", url, headers=headers, data=payload)
    return json.loads(response.text)["plainText"]

poemUrls = []
getPoemUrls('/bahar', poemUrls)
allPoems = ""
for poemUrl in poemUrls:
    allPoems = allPoems + getPoemByUrl(poemUrl) + "\r\n"
    f = open("bahar.sher", "w")
    f.write(allPoems)
    f.close()
    print(poemUrl)

```
